```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Set height of the bar
barWidth = 0.25
fig = plt.subplots(figsize = (12, 8))

#Set height of the bar
Games = [8, 8, 11, 6, 9, 8, 7, 10, 12, 9, 10]
Shots = [246, 298, 321, 177, 266, 195, 198, 265, 352, 322, 214]
Shots_Against = [235, 219, 340, 160, 242, 290, 178, 298, 379, 210, 303]

#Set postion of bar
br1 = np.arange(len(Games))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot
plt.bar(br1, Games, color ='r', width = barWidth,
        edgecolor ='grey', label ='Games')
plt.bar(br2, Shots, color ='g', width = barWidth,
        edgecolor ='grey', label ='Shots')
plt.bar(br3, Shots_Against, color ='b', width = barWidth,
        edgecolor ='grey', label ='Shots Against')
 
# Adding Xticks
plt.xlabel('Teams', fontweight ='bold', fontsize = 15)
plt.ylabel('Total Shots', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Games))],
        ['BC', 'BU', 'Connecticut', 'Maine', 'Mass-Lowell', 'Massachusetts', 'Merrimack', 'New Hampshire', 'Northeastern', 'Providence', 'Vermont'])
 
plt.legend()
plt.show()

```
